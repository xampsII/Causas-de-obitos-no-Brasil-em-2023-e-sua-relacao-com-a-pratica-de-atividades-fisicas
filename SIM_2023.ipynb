```python
df_idade_400_499 = df_filtrado[(df['IDADE'] >= 400) & (df_filtrado['IDADE'] <= 499)].copy()
```

```python
df_idade_400_499['IDADE_SEM_4'] = df_idade_400_499['IDADE'] % 100
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df_idade_400_499['SEXO'] = df_idade_400_499['SEXO'].map({1: 'Masculino', 2: 'Feminino'})
palette = {
    'Masculino': '#A2D9FF',  # Azul Claro
    'Feminino': '#A3D9A5'    # Verde Pastel
}

# Criando o gráfico de boxplot
plt.figure(figsize=(12, 8))

# Remover o fundo com linhas de grid
sns.set(style="white")  # Remover a grade de fundo (sem gridlines)

# Ajustando o gráfico para aplicar a paleta corretamente
sns.boxplot(x="SEXO", y="IDADE_SEM_4", data=df_idade_400_499, hue="SEXO", palette=palette, linewidth=1)

# Personalizando o gráfico
plt.title("Distribuição de Idade por Sexo", fontsize=16)
plt.xlabel("Sexo", fontsize=12)
plt.ylabel("Idade", fontsize=12)

plt.savefig("distribuicao_idade_por_sexo.png", format="png", dpi=600, bbox_inches='tight')  # dpi maior para mais qualidade


# Exibindo o gráfico
plt.show()
```

```python
# Criar faixas etárias de 10 em 10 anos
bins = range(0, df_idade_400_499['IDADE_SEM_4'].max() + 10, 10)  # ex: 0,10,20,30,...
labels = [f'{b}-{b+9}' for b in bins[:-1]]

df_idade_400_499['faixa_etaria'] = pd.cut(
    df_idade_400_499['IDADE_SEM_4'],
    bins=bins,
    labels=labels,
    right=False
)

# Calcular a moda do CID por faixa etária (sem warnings)
cid_moda_por_faixa = (
    df_idade_400_499
    .groupby('faixa_etaria', observed=False)['CAUSABAS']
    .agg(lambda x: x.mode().iloc[0])
    .reset_index()
)

# Plotar gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(cid_moda_por_faixa['faixa_etaria'].astype(str), cid_moda_por_faixa['CAUSABAS'])
plt.xticks(rotation=45)
plt.xlabel('Faixa Etária')
plt.ylabel('CID mais comum')
plt.title('CID mais comum por faixa etária')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
bars = plt.barh(cid_moda_por_faixa['faixa_etaria'].astype(str), range(len(cid_moda_por_faixa)))

# Adiciona o texto do CID dentro das barras
for i, (cid, bar) in enumerate(zip(cid_moda_por_faixa['CAUSABAS'], bars)):
    plt.text(bar.get_width()/2, bar.get_y() + bar.get_height()/2, cid, ha='center', va='center', color='white', fontsize=12)

plt.xlabel('Faixa Etária')
plt.title('CID mais comum por faixa etária')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Criar faixas etárias de 10 em 10 anos
bins = range(0, df_idade_400_499['IDADE_SEM_4'].max() + 10, 10)
labels = [f'{b}-{b+9}' for b in bins[:-1]]

df_idade_400_499['faixa_etaria'] = pd.cut(
    df_idade_400_499['IDADE_SEM_4'],
    bins=bins,
    labels=labels,
    right=False
)

# Contar as ocorrências de cada CID por faixa etária (sem o warning)
cid_por_faixa = (
    df_idade_400_499
    .groupby(['faixa_etaria', 'CAUSABAS'], observed=False)
    .size()
    .unstack(fill_value=0)
)

# Selecionar os 6 CIDs mais frequentes por faixa etária
def top_n_cids(row, n=6):
    return row.nlargest(n)

cid_por_faixa_top6 = cid_por_faixa.apply(top_n_cids, axis=1)

# Cores pastéis
cores_claras = [
    '#A2D9FF', '#A3D9A5', '#D1D8E0', '#FFA07A', '#D9A2FF',
    '#FFD9A2', '#A2F0FF', '#FFDAA2', '#A2FFD9', '#D1FFDA',
    '#F5CBA7', '#AED6F1', '#F9E79F', '#D7BDE2', '#A9DFBF',
    '#FADBD8', '#FDEBD0', '#D6EAF8', '#FDEDEC', '#E8DAEF',
    '#D5F5E3', '#FCF3CF', '#EBDEF0', '#F6DDCC', '#D4E6F1',
    '#F9EBEA', '#EAF2F8', '#FDEBD0', '#E5E8E8'
]

# Gráfico horizontal empilhado
ax = cid_por_faixa_top6.plot(
    kind='barh',
    stacked=True,
    figsize=(12, 8),
    color=cores_claras[:len(cid_por_faixa_top6.columns)]
)

plt.title('Distribuição dos 6 CIDs Mais Frequentes por Faixa Etária')
plt.xlabel('Número de Casos')
plt.ylabel('Faixa Etária')
plt.legend(title='CID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("distribuicao_CID_mais_frequentes.png", format="png", dpi=600, bbox_inches='tight')
plt.show()
```
